```python
def list_directories_and_files(account_name, storage_key, share_name, directory_name, file_or_dirs, recursive=True):
    file_service = get_file_service(account_name, storage_key)
    generator = file_service.list_directories_and_files(share_name, directory_name)
    for file_or_dir in generator:
        file_or_dirs.append(os.path.join(directory_name, file_or_dir.name))
        if recursive and not '.' in file_or_dir.name: # What is the File SDK equivalent of os.path.isfile?
            list_directories_and_files(account_name, storage_key, share_name, os.path.join(directory_name, file_or_dir.name), file_or_dirs)
    return file_or_dirs
```
